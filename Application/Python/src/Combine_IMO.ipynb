```python
files = [
    "FirstTrial_tanker_imos.txt",
    "SecondTrial_tanker_imos.txt",
    "ThirdTrial_tanker_imos.txt"
]

unique_lines = set()

for file in files:
    with open(file) as f:
        unique_lines.update(f.readlines())

with open("unique_rows.txt", "w") as out:
    out.writelines(unique_lines)

print("Done!")
```
